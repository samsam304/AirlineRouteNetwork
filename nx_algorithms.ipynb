## Algorithmic Analysis

The following functions are not included in the top level `networkx` namespace and must be imported independently.

To keep the focus on the functions and analysis instead of the graph generation, I will use the `watts_strogatz_graph(n,k,p)` function to generate random small world graphs. Where n is the number of node, k is the number of nearest neighbors it is joined with and p is the probability of rewiring each edge.

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.generators.random_graphs import watts_strogatz_graph

G=watts_strogatz_graph(10, 2, 0.5)
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

### Approximation and Heuristics

#### Connectivity
1. `all_paris_node_connectivity`(G[,nbunch cutoff]) &emsp; Computes connectivity between all pairs of nodes
2. `local_node_connectivity`(G,source,target[,...]) &emsp; Computes node connectivity between source and target
3. `node_connectivity`(G{,s,t]) &emsp; Returns an approximation for node connectivity for a graph or digraph G

#### Cliques
1. `max_clique`(G) &emsp; Find the maximum clique
2. `large_clique_size`(G) &emsp; Find the size of a large clique in a graph

#### Clustering
1. `average_clustering`(G) &emsp; Estimate average clustering coefficient

```python
from networkx.algorithms import approximation
```

```python
approximation.all_pairs_node_connectivity(G)
```

```python
approximation.large_clique_size(G)
```

```python
approximation.average_clustering(G)
```

```python
approximation.max_clique(G)
```

```python
approximation.large_clique_size(G)
```

```python
approximation.average_clustering(G)
```

### Assortativity

#### Node Degree

1. `average_neighbor_degree`(G) &emsp; Returns average degree of the neighborhood of each node
2. `average_degree_connectivity`(G) &emsp; Compute the average degree connectivity of the graph

```python
from networkx.algorithms import assortativity
```

```python
assortativity.average_neighbor_degree(G)
```

```python
assortativity.average_degree_connectivity(G)
```

### Centrality

#### Degree
1. `degree_centrality`(G) &emsp; Compute the degree centrality for nodes

#### Betweenness
1. `betweenness_centrality`(G) &emsp; Compute the shortest path betweenness centrality for nodes
2. `betweenness_centrality_subset`(G, sources) &emsp; Compute betweenness centrality for a subset of nodes



```python
from networkx.algorithms import centrality
```

```python
centrality.degree_centrality(G)
```

```python
centrality.betweenness_centrality(G)
```

```python
centrality.betweenness_centrality_subset(G,[4,9],[1])
```
